## Nilai Unik di Kolom Kategorikal

```python
df[['Department','EnvironmentSatisfaction','JobInvolvement','JobLevel','JobSatisfaction','PerformanceRating','RelationshipSatisfaction','WorkLifeBalance']] = df[['Department','EnvironmentSatisfaction','JobInvolvement','JobLevel','JobSatisfaction','PerformanceRating','RelationshipSatisfaction','WorkLifeBalance']].astype('str')
```

```python
# Jumlah kolom kategori
cat_columns = list(df.select_dtypes(include='object').columns)
n = len(cat_columns)
print(cat_columns)
print(n)
```

```python
# List untuk menyimpan tiap dataframe
tabel_kategori = []

for col in cat_columns:
    df_temp = df[col].value_counts().reset_index().rename(columns={
        'index': f'Nilai Unik ({col})',
        col: 'Jumlah'
    })
    # Sisipkan nama kolom & jumlah nilai unik sebagai metadata (optional)
    print(f"\nðŸ“Œ Kolom: {col}  â€” Jumlah nilai unik: {df[col].nunique()}")
    display(df_temp)  # Jika pakai Jupyter, tampil sebagai tabel
    tabel_kategori.append((col, df_temp))
```

## Batch Visualization of Categorical and Numerical Distributions via Subplots

### Categorical Data

```python
# Hitung ukuran grid yang pas
n_cols = 3  # Misal 3 kolom
n_rows = (n + n_cols - 1) // n_cols  # Pembulatan ke atas

# Buat subplot
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(n_cols * 5, n_rows * 4))
axes = axes.flatten()  # Ubah ke array 1D agar mudah di-loop

# Loop dan plot
for i, col in enumerate(cat_columns):
    df[col].value_counts().plot(kind='bar', ax=axes[i], title=col, color='salmon', edgecolor='black')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Jumlah')
    axes[i].tick_params(axis='x', rotation=90)

# Hapus axis sisa jika jumlah kolom tidak pas
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### Numerical Data

```python
num_columns = list(df.select_dtypes(include=['int64', 'float64']).columns)
n = len(num_columns)
print(num_columns)
print(n)
```

```python
n_cols = 4  # Misal 3 kolom
n_rows = (n + n_cols - 1) // n_cols  # Pembulatan ke atas

# Buat subplot
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(n_cols * 5, n_rows * 4))
axes = axes.flatten()  # Ubah ke array 1D agar mudah di-loop

# Loop dan plot
for i, col in enumerate(num_columns):
    df[col].value_counts().plot(kind='hist', ax=axes[i], title=col)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Jumlah')
    axes[i].tick_params(axis='x', rotation=-45)

# Hapus axis sisa jika jumlah kolom tidak pas
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# `Attrition` Analysis

```python
cat_columns_targetless = df.select_dtypes(include='object').columns.drop('Attrition')  # Kecuali target

# List hasil
ringkasan_attrition = []

for col in cat_columns:
    # Hitung distribusi Attrition untuk setiap kategori
    tab = pd.crosstab(df[col], df['Attrition'], margins=True)
    
    # Hitung rasio Attrition 'Yes' terhadap total tiap kategori
    tab['Attrition_Yes_Ratio (%)'] = round(tab['Yes'] / tab['All'] * 100, 2)
    
    # Hapus baris 'All' agar tidak tercampur
    tab = tab.drop(index='All')
    
    # Reset index agar rapi ditampilkan
    tab = tab.reset_index()
    
    print(f"\nðŸ“Œ Ringkasan Attrition berdasarkan: {col}")
    display(tab)
    
    # Simpan jika ingin digunakan kembali
    ringkasan_attrition.append((col, tab))
```

# Export to Tableau

```python
# Salin dataframe asli
df_new = df.copy()

# Mapping dictionary berdasarkan deskripsi Kaggle
edu_map = {
    '1': 'Below College', '2': 'College', '3': 'Bachelor', '4': 'Master', '5': 'Doctor'
}

satisfaction_map = {
    '1': 'Low', '2': 'Medium', '3': 'High', '4': 'Very High'
}

performance_map = {
    '1': 'Low', '2': 'Good', '3': 'Excellent', '4': 'Outstanding'
}

wlb_map = {
    '1': 'Bad', '2': 'Good', '3': 'Better', '4': 'Best'
}

# Terapkan mapping
df_new['Education'] = df_new['Education'].map(edu_map)
df_new['EnvironmentSatisfaction'] = df_new['EnvironmentSatisfaction'].map(satisfaction_map)
df_new['JobInvolvement'] = df_new['JobInvolvement'].map(satisfaction_map)
df_new['JobSatisfaction'] = df_new['JobSatisfaction'].map(satisfaction_map)
df_new['PerformanceRating'] = df_new['PerformanceRating'].map(performance_map)
df_new['RelationshipSatisfaction'] = df_new['RelationshipSatisfaction'].map(satisfaction_map)
df_new['WorkLifeBalance'] = df_new['WorkLifeBalance'].map(wlb_map)

# Export ke CSV
df_new.to_csv("/IBM HR Analytics Employee Attrition and Performance/data_ibm_hr_for_tableau.csv", index=False)

print("âœ… Dataset berhasil disiapkan dan diekspor ke 'data_ibm_hr_for_tableau.csv'")
```
